```python
# importar a biblioteca
import pandas as pd
```

```python
# leitura dos arquivos
df1 = pd.read_excel("./datasets/Aracaju.xlsx")
df2 = pd.read_excel("./datasets/Fortaleza.xlsx")
df3 = pd.read_excel("./datasets/Natal.xlsx")
df4 = pd.read_excel("./datasets/Recife.xlsx")
df5 = pd.read_excel("./datasets/Salvador.xlsx")
```

```python
display(df5)
```

```python
# juntar os arquivo um baixo do outro
df = pd.concat([df1, df2, df3, df4, df5])
df.info()
```

```python
# pegar um amostrada da base
df.sample(5)
```

```python
# tipo de dado
df.dtypes
```

```python
# Alterar o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
df.dtypes
```

### Tratamento de valores faltantes

```python
# Consultando linhas com valores faltantes (quatidade)
df.isnull().sum()
```

```python
# Substituindo os valores nulos pela mÃ©dia
df["Vendas"].fillna(df["Vendas"].mean(), implace=True)
```

```python
# Substituir os valores nulos por zeros
df["Vendas"].fillna(0, inplace=True)
```

```python
# Apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
# Apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
# Removendo linas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace="True")
```

### Criando colunas novas

```python
df.sample(15)
```

```python
# Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
# Criando um coluna pela divisÃ£o de outras colunas
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
# Retornar a maior receita
df["Receita"].max()
```

```python
# Retornando a menor receita
df["Receita"].min()
```

```python
# nlargest - as 3 linhas com maior receita
df.nlargest(3, "Receita")
```

```python
# nsmallest - as 3 inhas com a menor receita
df.nsmallest(3, "Receita")
```

```python
# agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
# ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```
